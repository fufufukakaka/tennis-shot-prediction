```python
##サーブに関する文言を切り取る関数
def deal_serve_words(serve_words):
    ##faultがあるかどうか
    ends = "Continue"
    if serve_words.find("Fault") >= 0:
        fault = 1
        serve0 = serve_words.split(".")[1][3:]
        ## ダブルフォルト？
        if serve0.find("DoubleFault") >= 0:
            serve = serve0.split(",")[0]
            ends = "DoubleFault"
        else:
            serve = serve0
    else:
        fault = 0
        
        ##エース？
        if serve_words.find("Ace") >= 0:
            serve = serve_words.split(",")[0][3:]
            ends = "Ace"
        else:
            serve = serve_words[3:]
    return fault,serve,ends
```

```python
deal_serve_words("1stServeWide,Fault(net).2ndServeToBody")
```

```python
##depthに関する文言を切り取る関数
def deal_depth(shot_words):
    if shot_words.find("shallow") >= 0 or shot_words.find("deep") >= 0 or shot_words.find("veryDeep") >= 0 or shot_words.find("netCord") >= 0:
        shots = shot_words.split("(")[0]
        depth = shot_words.split("(")[1].replace(")","")
    else:
        shots = shot_words
        depth = "Normal"
    return shots,depth
```

```python
deal_depth("ForehandReturnInside-out(veryDeep)")
```

```python
##ラリー終了に関する文言を切り取る関数
def deal_ends(shot_words):
    if shot_words.find("Error") >= 0:
        shot_errors = shot_words.split(",")
        if shot_errors[0].find("(") >= 0:
            shots = shot_errors[0].split("(")[0]
            error_direction = shot_errors[0].split("(")[1].replace(")","")
            err0 = shot_errors[1].replace(".","")
            if err0.find("Rally") >= 0:
                ends = error_direction.capitalize() + err0.split("(")[0]
            else:
                ends = error_direction.capitalize() + err0
        else:
            ##No Direction
            shots = shot_errors[0]
            err0 = shot_errors[1].replace(".","")
            if err0.find("Rally") >= 0:
                ends = err0.split("(")[0]
            else:
                ends = err0
    else:
        shots = shot_words.split(",")[0]
        ends0 = shot_words.split(",")[1].replace(".","")
        if ends0.find("Rally") >= 0:
            ends = ends0.split("(")[0]
        else:
            ends = ends0
    return shots,ends
```

```python
from tqdm import tqdm_notebook
def preprocess_shot_detail(X_array,end_array,depth_array,fault_array,shooter_array,set_array2,game_array2,point_array2,server_array2,shot_array,set_array,game_array,point_array,server_array,federer_position_array,nadal_position_array,federer_position_array2,nadal_position_array2):
    for shot,sets,game,point,server,federer_p,nadal_p in tqdm_notebook(zip(shot_array,set_array,game_array,point_array,server_array,federer_position_array,nadal_position_array)):
        ##最初にボールを打つのはサーバー
        shooter = server
        for i in range(len(shot)):
            s_array = []
            fp_array = []
            np_array = []
            s = shot[i]
            try:
                fp = federer_p[i]
                np = nadal_p[i]
            except:
                ##記入がミスっていた場合は最後のポジションを使う
                fp = federer_p[-1]
                np = nadal_p[-1]
            depth = 'Normal'
            ends = "Continue"
            fault = 0
            ## サーブの場合の文字列処理
            if i == 0:
                fault,s_detail,ends = deal_serve_words(s)
            elif i < len(shot)-1:
                ## depthに関する文字列処理
#                 if s.find("shallow") >= 0 or s.find("deep") >= 0 or s.find("veryDeep") >= 0 or s.find("netCord") >= 0:
#                     print(s)
                s_detail,depth = deal_depth(s)
#                 if s.find("shallow") >= 0 or s.find("deep") >= 0 or s.find("veryDeep") >= 0 or s.find("netCord") >= 0:
#                     print(s_detail,depth)
            ## ラリー終了に関する文字列処理
            elif i == len(shot)-1:
                s0,ends = deal_ends(s)
                s_detail,depth = deal_depth(s0)
            
            #from IPython.core.debugger import Pdb; Pdb().set_trace()
            ## ラリーについて
#             print('X_array {}'.format(X_array))
            if i == 0:
                s_array.append(s_detail)
            else:
                s_tmp = X_array[-1].copy()
#                 print('s_tmp {}'.format(s_tmp))
#                 print('s_array {}'.format(s_array))
                s_tmp.append(s_detail)
                s_array = s_tmp
                #from IPython.core.debugger import Pdb; Pdb().set_trace()
            
#             print('s_detail {}'.format(s_detail))
#             print('s_array {}'.format(s_array))
            X_array.append(s_array)
            end_array.append(ends)
            depth_array.append(depth)
            fault_array.append(fault)
            set_array2.append(sets)
            game_array2.append(game)
            point_array2.append(point)
            server_array2.append(server)

            ##ショットを打っているのはどちらか
            shooter_array.append(shooter)
            if shooter == "Federer":
                shooter = "Nadal"
            elif shooter == "Nadal":
                shooter = "Federer"
                
            ##ポジション
            if i == 0:
                fp_array.append(fp)
                np_array.append(np)
            else:
                fp_tmp = federer_position_array2[-1].copy()
                fp_tmp.append(fp)
                fp_array = fp_tmp
                np_tmp = nadal_position_array2[-1].copy()
                np_tmp.append(np)
                np_array = np_tmp
            federer_position_array2.append(fp_array)
            nadal_position_array2.append(np_array)
                
    return X_array,end_array,depth_array,fault_array,shooter_array,set_array2,game_array2,point_array2,server_array2,federer_position_array2,nadal_position_array2
```

```python
def preprocess_shot_details(match_excel,match_name):
    df = pd.read_excel(match_excel)
    
    ##第一段階処理
    shot_array = []
    set_array = []
    game_array = []
    point_array = []
    server_array = []
    federer_position_array = []
    nadal_position_array = []
    for idx,row in df.iterrows():
        shots = row["Detail"].split("; ")
        s_array = []
        for s in shots:
            s_array.append("".join([v.capitalize() for v in s.split(" ")]))
        shot_array.append(s_array)

        set_array.append(row["Sets"])
        game_array.append(row["Games"])
        point_array.append(row["Points"])
        server_array.append(row["Server"].split(" ")[1])

        ##ポジション
        federer_position_array.append(str(row["FedererPosition"]).split(","))
        nadal_position_array.append(str(row["NadalPosition"]).split(","))
    
    ##第二段階処理
    X_array = []
    end_array = []
    depth_array = []
    fault_array = []
    shooter_array = []
    set_array2 = []
    game_array2 = []
    point_array2 = []
    server_array2 = []
    federer_position_array2 = []
    nadal_position_array2 = []
    X_array, end_array, depth_array, fault_array, shooter_array, set_array2, game_array2, point_array2, server_array2, federer_position_array2, nadal_position_array2 = preprocess_shot_detail(
        X_array, end_array, depth_array, fault_array, shooter_array,
        set_array2, game_array2, point_array2, server_array2, shot_array,
        set_array, game_array, point_array, server_array, federer_position_array,
        nadal_position_array, federer_position_array2, nadal_position_array2)
    
    ##第三段階
    rally_df = pd.DataFrame([])
    rally_df["set"] = set_array2
    rally_df["game"] = game_array2
    rally_df["point"] = point_array2
    rally_df["server"] = server_array2
    rally_df["shooter"] = shooter_array
    rally_df["rally"] = X_array
    rally_df["end"] = end_array
    rally_df["depth"] = depth_array
    rally_df["fault"] = fault_array
    rally_df["federer_position"] = federer_position_array2
    rally_df["nadal_position"] = nadal_position_array2
    
    ## Xとyに分ける
    rally_array = list(rally_df["rally"])
    X = []
    y = []
    for r in rally_array:
        if len(r) == 1:
            ##サーブだけの行はあとでdropする
            xt = r
            yt = "TMP"
        else:
            ##末尾がyになる
            yt = r[-1]
            r1 = r.copy()
            r1.pop()
            ##末尾を除いたものがxになる
            xt = r1
        X.append(xt)
        y.append(yt)
        
    rally_df["X"] = X
    rally_df["y"] = y
    rally_df_drop = rally_df.loc[rally_df["y"]!="TMP"]
    rally_df_drop = rally_df_drop[['set', 'game', 'point', 'server', 'shooter', 'rally', 'depth',
       'fault', 'federer_position', 'nadal_position', 'X', 'end', 'y']]
    rally_df_drop["match"] = match_name
    
    return rally_df_drop
```

```python
rdf1 = preprocess_shot_details("../data/excel/federerVsNadalAustralianOpen2012.xlsx","federerVsNadalAustralianOpen2012")
rdf2 = preprocess_shot_details("../data/excel/federerVsNadalAustralianOpen2014.xlsx","federerVsNadalAustralianOpen2014")
rdf3 = preprocess_shot_details("../data/excel/federerVsNadalRolandGarros2007.xlsx","federerVsNadalRolandGarros2007")
rdf4 = preprocess_shot_details("../data/excel/federerVsNadalWimbledon2008.xlsx","federerVsNadalWimbledon2008")
```
