```python
##サーブに関する文言を切り取る関数
def deal_serve_words(serve_words):
    ##faultがあるかどうか
    ends = "Continue"
    if serve_words.find("Fault") >= 0:
        fault = 1
        serve0 = serve_words.split(".")[1][3:]
        ## ダブルフォルト？
        if serve0.find("DoubleFault") >= 0:
            serve = serve0.split(",")[0]
            ends = "DoubleFault"
        else:
            serve = serve0
    else:
        fault = 0
        
        ##エース？
        if serve_words.find("Ace") >= 0:
            serve = serve_words.split(",")[0][3:]
            ends = "Ace"
        else:
            serve = serve_words[3:]
    return fault,serve,ends
```

```python
deal_serve_words("1stServeWide,Fault(net).2ndServeToBody")
```

```python
##depthに関する文言を切り取る関数
def deal_depth(shot_words):
    if shot_words.find("shallow") >= 0 or shot_words.find("deep") >= 0 or shot_words.find("veryDeep") >= 0 or shot_words.find("netCord") >= 0:
        shots = shot_words.split("(")[0]
        depth = shot_words.split("(")[1].replace(")","")
    else:
        shots = shot_words
        depth = "Normal"
    return shots,depth
```

```python
deal_depth("ForehandReturnInside-out(veryDeep)")
```

```python
##ラリー終了に関する文言を切り取る関数
def deal_ends(shot_words):
    if shot_words.find("Error") >= 0:
        shot_errors = shot_words.split(",")
        if shot_errors[0].find("(") >= 0:
            shots = shot_errors[0].split("(")[0]
            error_direction = shot_errors[0].split("(")[1].replace(")","")
            err0 = shot_errors[1].replace(".","")
            if err0.find("Rally") >= 0:
                ends = error_direction.capitalize() + err0.split("(")[0]
            else:
                ends = error_direction.capitalize() + err0
        else:
            ##No Direction
            shots = shot_errors[0]
            err0 = shot_errors[1].replace(".","")
            if err0.find("Rally") >= 0:
                ends = err0.split("(")[0]
            else:
                ends = err0
    else:
        shots = shot_words.split(",")[0]
        ends0 = shot_words.split(",")[1].replace(".","")
        if ends0.find("Rally") >= 0:
            ends = ends0.split("(")[0]
        else:
            ends = ends0
    return shots,ends
```

```python
deal_ends('Forehand, ForcedError(5-shotRally)')
```

```python
from tqdm import tqdm_notebook
def preprocess_shot_detail(X_array,y_array,end_array,depth_array,fault_array,shooter_array,set_array2,game_array2,point_array2,server_array2,shot_array,set_array,game_array,point_array,server_array):
    for shot,sets,game,point,server in tqdm_notebook(zip(shot_array,set_array,game_array,point_array,server_array)):
        ##最初にボールを打つのはサーバー
        shooter = server
        for i in range(len(shot)):
            s_array = []
            s = shot[i]
            depth = 'Normal'
            ends = "Continue"
            fault = 0
            ## サーブの場合の文字列処理
            if i == 0:
                fault,s_detail,ends = deal_serve_words(s)
            elif i < len(shot)-1:
                ## depthに関する文字列処理
#                 if s.find("shallow") >= 0 or s.find("deep") >= 0 or s.find("veryDeep") >= 0 or s.find("netCord") >= 0:
#                     print(s)
                s_detail,depth = deal_depth(s)
#                 if s.find("shallow") >= 0 or s.find("deep") >= 0 or s.find("veryDeep") >= 0 or s.find("netCord") >= 0:
#                     print(s_detail,depth)
            ## ラリー終了に関する文字列処理
            elif i == len(shot)-1:
                s0,ends = deal_ends(s)
                s_detail,depth = deal_depth(s0)
            
            #from IPython.core.debugger import Pdb; Pdb().set_trace()
            ## ラリーについて
#             print('X_array {}'.format(X_array))
            if i == 0:
                s_array.append(s_detail)
            else:
                s_tmp = X_array[-1].copy()
#                 print('s_tmp {}'.format(s_tmp))
#                 print('s_array {}'.format(s_array))
                s_tmp.append(s_detail)
                s_array = s_tmp
                #from IPython.core.debugger import Pdb; Pdb().set_trace()
            
#             print('s_detail {}'.format(s_detail))
#             print('s_array {}'.format(s_array))
            X_array.append(s_array)
            end_array.append(ends)
            depth_array.append(depth)
            fault_array.append(fault)
            set_array2.append(sets)
            game_array2.append(game)
            point_array2.append(point)
            server_array2.append(server)

            ##ショットを打っているのはどちらか
            shooter_array.append(shooter)
            if shooter == "Federer":
                shooter = "Nadal"
            elif shooter == "Nadal":
                shooter = "Federer"
                
    return X_array,y_array,end_array,depth_array,fault_array,shooter_array,set_array2,game_array2,point_array2,server_array2
```

```python
X_array = []
y_array = []
end_array = []
depth_array = []
fault_array = []
shooter_array = []
set_array2 = []
game_array2 = []
point_array2 = []
server_array2 = []
X_array,y_array,end_array,depth_array,fault_array,shooter_array,set_array2,game_array2,point_array2,server_array2 = preprocess_shot_detail(X_array,y_array,end_array,depth_array,fault_array,shooter_array,set_array2,game_array2,point_array2,server_array2,shot_array,set_array,game_array,point_array,server_array)
```
